## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

```python
import pandas as pd
import numpy as np
import statistics as stats
import seaborn as sns
```

```python
customer = pd.read_csv('../lab-dw-pandas/customer.csv')
customer
```

```python
#convert titles to lower case and replace spaces with_

columns = []
for col in customer.columns: 
    if ' ' in col:
        col = col.replace(' ','_')
        columns.append(col.lower())
    else:
        columns.append(col.lower())
        
customer.columns = columns
customer
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they contain only valid data. For example, the gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

```python
#We only want F and M so let's change the rest to make it uniform.
customer['gender'].value_counts()
```

```python
#we make a loop where errors are converted to M or F

genderok = []
for i in customer['gender']:
    #if Male append M
    if i =='Male':
        genderok.append('M')
    #elif female append F
    elif i =='female':
        genderok.append('F')
    #elif Femal append F
    elif i =='Femal':
        genderok.append('F')
    #else append i
    else:
        genderok.append(i)

#The 'gender' column now has the data of genderok
customer['gender'] = genderok
```

## Exercise 3: Formatting data types

The data types of many columns in the dataset appear to be incorrect. This could impact the accuracy of our analysis. To ensure accurate analysis, we need to correct the data types of these columns. Please update the data types of the columns as appropriate. 

It is important to note that this exercise does not involve handling null values (NaN or None).

```python
#Change Customer Lifetime Value column to be a float. Adding % to column name. 

customer['customer_lifetime_value'] = customer['customer_lifetime_value'].str.rstrip('%').astype('float') / 100.0
customer.rename(columns = {'customer_lifetime_value':'customer_lifetime_value_%'}, inplace = True)

customer['customer_lifetime_value_%'] = round(customer['customer_lifetime_value_%'],2)
```

```python
#In column Income and in Monthly Premium Auto,looks like all the floats are .0, we change to integers.

customer['income'] = pd.to_numeric(customer['income'])
customer['monthly_premium_auto'] = pd.to_numeric(customer['monthly_premium_auto'])
```

```python
#In column Number of Open Complaints, looks like an integer is showed like a date. 1/00/00 should be 1. We change to an integer.

customer['number_of_open_complaints'] = customer['number_of_open_complaints'].str.split('/',n=2, expand=True)[1]
```

```python
#In column Total Claim Amount, floats are too long. We change to just the first two decimal places. 

customer['total_claim_amount'] = round(customer['total_claim_amount'], 2)
```

```python
# check remaining nans
customer.isna().sum()
```

```python
# 5. convert all the numeric variables to integers using applymap. 

cols = customer.select_dtypes(include=np.number).columns
customer[cols] = customer[cols].applymap(np.int64)
customer
```

# Challenge 2: creating functions on a separate `py` file

Put all the data cleaning and formatting steps into functions, and create a main function that performs all the cleaning and formatting. 

Write these functions in a separate .py file(s). By putting these steps into functions, we can make the code more modular and easier to maintain.

```python
# turn column headers into snake_case
def column_snake_case(x):
    columns = []
    for col in x.columns: 
        if ' ' in col:
            col = col.replace(' ','_')
            columns.append(col.lower())
        else:
            columns.append(col.lower())
            
    x.columns = columns
    return x.columnns

# make percentage objs into floats for manipulation
def percentage_to_float(x):
    x = x.str.rstrip('%').astype('float') / 100.0
    x = round(x,2)
    return x

# clean and format genders
def genderok(x):
    genderok = []
    for i in x['gender']:
        #if Male append M
        if i =='Male':
            genderok.append('M')
        #elif female append F
        elif i =='female':
            genderok.append('F')
        #elif Femal append F
        elif i =='Femal':
            genderok.append('F')
        #else append i
        else:
            genderok.append(i)
    
    #The 'gender' column now has the data of genderok
    x['gender'] = genderok
    return x['gender']
```
